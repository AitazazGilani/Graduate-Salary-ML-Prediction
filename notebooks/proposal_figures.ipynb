```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

np.random.seed(0)

# Generate random GPA data between 1.0 and 4.0
gpa = np.random.uniform(1.0, 4.0, 100)

# Create a binary outcome variable based on GPA
income = (gpa + np.random.normal(0, 0.5, 100)) > 2.5  # Adjust the threshold (e.g., 2.5) as needed

# logistic regression model
model = LogisticRegression()
model.fit(gpa.reshape(-1, 1), income)

# Generate a range of GPA values for the X-axis
gpa_range = np.linspace(1.0, 4.0, 100).reshape(-1, 1)

predicted_probabilities = model.predict_proba(gpa_range)[:, 1]

# Create a scatter plot of the dummy data points
plt.scatter(gpa, income, color='skyblue', label='Data Points')

# Plot the logistic regression curve
plt.plot(gpa_range, predicted_probabilities, color='red', label='Logistic Regression Curve')

# Set labels and title
plt.xlabel('GPA')
plt.ylabel('Probability of a High Income Career')
plt.title('Logistic Regression: GPA vs. High Income Prediction')


plt.legend()
plt.show()
```
